### Objective:

##### Analyze the Gold-layer datasets to uncover business insights related to channel performance, country-wise engagement, category trends, upload timing optimization, and top-performing videos. The results of this notebook are used for reporting and Power BI dashboard development.

#### Read Gold Tables

```python
top_channels = spark.read.format("delta").load(
    "/Volumes/workspace/default/youtube_data/gold_top_channels"
)

country_performance = spark.read.format("delta").load(
    "/Volumes/workspace/default/youtube_data/gold_country_performance"
)

category_performance = spark.read.format("delta").load(
    "/Volumes/workspace/default/youtube_data/gold_category_performance"
)

upload_time = spark.read.format("delta").load(
    "/Volumes/workspace/default/youtube_data/gold_upload_time"
)

top_videos = spark.read.format("delta").load(
    "/Volumes/workspace/default/youtube_data/gold_top_videos"
)
```

#### Analysis 1: Top 10 Channels

```python
display(
    top_channels.orderBy(
        "total_views",
        ascending=False
    ).limit(10)
)
```

#### Analysis 2: Best Performing Country

```python
display(
    country_performance.orderBy(
        "total_views",
        ascending=False
    )
)
```

#### Analysis 3: Most Popular Categories

```python
display(
    category_performance.select(
        "category_name",
        "category_id",
        "total_videos",
        "total_views",
        "avg_engagement_rate"
    ).orderBy(
        "total_views",
        ascending=False
    )
)
```

#### Analysis 4: Best Upload Time

```python
display(
    upload_time.orderBy(
        "avg_views",
        ascending=False
    )
)
```

#### Analysis 5: Most Engaging Videos

```python
display(
    top_videos.orderBy(
        "engagement_rate",
        ascending=False
    ).limit(20)
)
```

#### Analysis 6: Country vs Engagement

```python
display(
    country_performance.orderBy(
        "avg_engagement_rate",
        ascending=False
    )
)
```

#### Analysis 7: Trending Videos Distribution

```python
display(
    country_performance.select("country", "total_videos")
    .orderBy("total_videos", ascending=False)
)
```

#### Analysis 8: Channels with Highest Engagement

```python
display(
    top_channels.select(
        "channel_title",
        "avg_engagement_rate"
    ).orderBy(
        "avg_engagement_rate",
        ascending=False
    ).limit(20)
)
```

#### Analysis 9: Monthly Trends

```python
from pyspark.sql.functions import sum

silver_df = spark.read.format("delta").load("/Volumes/workspace/default/youtube_data/silver")

display(
    silver_df.groupBy(
        "publish_month"
    ).agg(
        sum("views").alias("total_views")
    ).orderBy(
        "publish_month"
    )
)
```

#### Analysis 10: Data Quality Report

```python
print("Total Records:", silver_df.count())

print(
    "Unique Channels:",
    silver_df.select("channel_title")
    .distinct()
    .count()
)

print(
    "Unique Categories:",
    silver_df.select("category_id")
    .distinct()
    .count()
)
```

#### Export Data for Power BI

```python
top_channels.toPandas().to_csv("top_channels.csv", index=False)
top_videos.toPandas().to_csv("top_videos.csv", index=False)
country_performance.toPandas().to_csv("country_performance.csv", index=False)
category_performance.toPandas().to_csv("category_performance.csv", index=False)
upload_time.toPandas().to_csv("upload_time.csv", index=False)
```
